```python
# Problem 1: Confusion Matrix
# Example: (TP=50, FP=8, TN=32, FN=10)

TP, FP, TN, FN = 50, 8, 32, 10
total = TP + FP + TN + FN  # Total predictions
accuracy = (TP + TN) / total
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1 = 2 * precision * recall / (precision + recall)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
# Problem 2: R²

ssr = 420  # Sum of Squares due to Regression
sse = 180  # Sum of Squares due to Error

# Total Sum of Squares (SST) = SSR + SSE
sst = ssr + sse

# R² = SSR / SST
r2 = ssr / sst

# Print results
print("SST:", sst)
print("R²:", r2)

print("The model explains", r2 * 100, "% of the variance.")
if r2 > 0.7:
    print("Yes, more than 70% of the variance is explained.")
else:
    print("No, it does not explain more than 70% of the variance.")
```

```python
# Problem 3: Mean Squared Error
# Example if values were [2,4,1,9,4]

sq_err1 = 2
sq_err2 = 4
sq_err3 = 1
sq_err4 = 9
sq_err5 = 4
sum_sq_errors = sq_err1 + sq_err2 + sq_err3 + sq_err4 + sq_err5
mse = sum_sq_errors / 5

# Print results
print("MSE:", mse)
```

```python
# Problem 4: Cook's Distance

e2 = 9         # Squared residual
mse = 3        # Mean Squared Error
p = 3          # Number of predictors
hii = 0.2      # Leverage value
n = 25         # Sample size

# Compute Cook’s Distance:
cook_d = (e2 * hii) / (p * mse * (1 - hii)**2)

# Common threshold:
threshold = 4 / (n - p - 1)

# Print results:
print("Cook’s Distance:", cook_d)
print("Threshold (4/(n-p-1)):", threshold)
if cook_d > threshold:
    print("The Observation is Influential.")
else:
    print("The Bbservation is not Influential.")
```
